```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
from astropy import units as u
from astropy import constants as const
from astropy.units import imperial
imperial.enable()
import nbconvert as nbconvert
data_table = ascii.read('E3Stetson.phot')
data_table2 = ascii.read('E3HST.phot')

Vg = data_table['col10']
Ig = data_table['col18']
Vs = data_table2['HST']
Is = data_table2['Photometry']
```

```python
%matplotlib inline

plt.figure(figsize=(20,15))
VI_mag = ((Vg - Ig) - .46)

plt.scatter(VI_mag, Vg, s=15);

plt.title("E3 Color-Magnitude Diagram (Stetson)", size = 20)

plt.xlabel("Ground photometry corrected", size = 20)
plt.ylabel("V-Ground", size = 20)

plt.xlim(-0.5,3)
plt.gca().invert_yaxis()
plt.text(0.75, 21, 'Sun', fontsize=20)
plt.plot(0.75, 21, marker='o', markersize=6, color="red")
plt.text(-0.5,22, 'Hotter', fontsize=12, color='blue')
plt.text(3,22, 'Cooler', fontsize=12, color='red')
```

```python
%matplotlib inline

plt.figure(figsize=(20,15))
VI_mag2 = ((Vs - Is) - .46)

plt.scatter(VI_mag2, Vs, s=9);

plt.title("E3 Color-Magnitude Diagram (HST)", size = 20)

plt.xlabel("HST photometry corrected", size = 20)
plt.ylabel("V-HST", size = 20)

plt.xlim(-2,5)
plt.gca().invert_yaxis()
plt.text(-2,31, 'Hotter', fontsize=12, color='blue')
plt.text(5,31, 'Cooler', fontsize=12, color='red')

```
